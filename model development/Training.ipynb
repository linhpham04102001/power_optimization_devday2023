# Assisting functions

```python
def rmr_score(actual, predict):
    """データフレームから残差平均割合を計算する。

    Args:
        actual (np.array): 実績
        predict (np.array): 予測

    Returns:
        float: 残差平均割合
    """
    eps = 1e-9
    actual = actual + eps
    diff = actual - predict


    mx = sum(abs(diff)) / sum(actual)
    return mx * 100
```

```python
def make_dataset_dl(_dataset, input_size, output_size, input_features, output_feature, shift=48, stride=48):
        dataset = _dataset.copy()
        window_size = input_size + shift + output_size
        input_list, output_list = None, None
        for i in range(0, len(dataset) - window_size + 1, stride):
            inp_s = i
            inp_e = i + input_size
            outp_s = i + input_size + shift
            outp_e = i + input_size + shift + output_size
            if type(input_list) is type(None):
                input_list = np.array([dataset[input_features].iloc[inp_s: inp_e].values])
            else:
                input_list = np.append(input_list, [dataset[input_features].iloc[inp_s: inp_e].values], axis=0)

            if type(output_list) is type(None):
                output_list = np.array([dataset[output_feature].iloc[outp_s: outp_e]])
            else:
                output_list = np.append(output_list, [dataset[output_feature].iloc[outp_s: outp_e]], axis=0)
        return input_list.astype(np.float32), output_list.astype(np.float32)
```

```python
from torch import nn
import torch.nn.functional as F
import torch
from tqdm import tqdm
class RNN_Net(nn.Module):
    def __init__(self, input_size, hidden_dim, num_layers, output_len=48, output_size=1, mode='LSTM', dropout=0, task='classification'):
        super(RNN_Net, self).__init__()
        self.hidden_dim = hidden_dim
        self.task = task
        self.output_len = output_len
        self.cell = nn.LSTM(input_size=input_size, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True, dropout=dropout)
        if mode == 'GRU':
            self.cell = nn.GRU(input_size=input_size, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True, dropout=dropout)
        elif mode == 'RNN':
            self.cell = nn.RNN(input_size=input_size, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.linear1 = nn.Linear(hidden_dim, 128)
        self.linear2 = nn.Linear(128, 32)
        self.linear3 = nn.Linear(32, output_size)

    def forward(self, x):           # x shape: (batch_size, seq_len, input_size)
        out, _ = self.cell(x)
        out = out[:,-self.output_len:,:]
        # if len(out.shape) == 2:
        #     out = out[-1,:]
        # else:
        #     out = out[:,-1,:]

        # out = out.reshape(-1, self.hidden_dim)

        out = F.relu(self.linear1(out))  # out shape: (batch_size, n_class=1)
        out = F.relu(self.linear2(out))
        out = self.linear3(out)


        if self.task == 'classification':
            out = F.sigmoid(out)
        return out
```

```python
from torch.utils.data import Dataset, DataLoader
class PowerDataset(Dataset):
    def __init__(self, _input, _output):
        self._input = _input
        self._output = _output

    def __len__(self):
        return len(self._input)

    def __getitem__(self, idx):
        return self._input[idx], self._output[idx]
```

```python
from sklearn.metrics import mean_squared_error
def train(model, train_loader, val_loader, output_scaler=None, out_dir='.', EPOCH=1000, seed=0, LR=0.001, weight_decay=0, device='cpu', clipping_value=None, verbose=100, task='classification'):
        if not os.path.isdir(out_dir):
            os.mkdir(out_dir)

        if type(val_loader) is type(None):
            print("no validation test --> no use early stopping")

        loss_list, y_ = [], []
        val_loss_list = []
        rmse_list, mape_list = [], []
        best_rmse = None
        best_val_loss= None
        model = model.to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=weight_decay)
        if task == 'classification':
            criterion = nn.BCELoss()
        else:
            criterion = nn.MSELoss()

        for epoch in tqdm(range(EPOCH)) :
            model.train()
            train_loss = []
            for x,y in train_loader:
                x = x.to(device)
                y = y.to(device)
                output = model(x)
                # print('y', y)
                # print('predict', output)
                loss = criterion(output, y)
                train_loss.append(loss.cpu().detach().numpy())
                optimizer.zero_grad()
                loss.backward()
                if clipping_value:
                    nn.utils.clip_grad_norm(model.parameters(), clipping_value)
                optimizer.step()

            if epoch%verbose==0 or epoch==(EPOCH-1):
                loss_list.append(np.mean(train_loss))
                if val_loader:
                    model.eval()
                    val_loss = []
                    for x, y in val_loader:
                        b = x.size(0)
                        x = x.to(device)
                        with torch.no_grad():
                            pred = model(x).cpu().detach().numpy().reshape((b, -1))
                        y = y.to('cpu').detach().numpy().reshape((b, -1))
                        if output_scaler:
                          pred = output_scaler.inverse_transform(pred)
                          y = output_scaler.inverse_transform(y)
                        pred = pred.flatten()
                        y = y.flatten()
                        # val_loss.append(mean_squared_error(y, pred))
                        # val_loss.append(BinaryCrossEntropy(y, pred))
                        val_loss.append(rmr_score(y, pred))
                    val_loss_list.append(np.mean(val_loss))

                    if best_val_loss == None:
                        best_val_loss = {}
                        best_val_loss['name'] = 'checkpoint.pt'
                        best_val_loss['value'] = val_loss_list[-1]
                        torch.save(model, os.path.join(out_dir,best_val_loss['name']))
                    else:
                        # if val_loss_list[-1] < best_val_loss['value']:
                        if val_loss_list[-1] < best_val_loss['value']:
                            os.remove(os.path.join(out_dir,best_val_loss['name']))
                            best_val_loss['name'] = 'checkpoint.pt'
                            best_val_loss['value'] = val_loss_list[-1]
                            torch.save(model, os.path.join(out_dir,best_val_loss['name']))

                    print('epoch:{:<2d} | loss:{:<6.7f} | val_loss:{:<6.7f}'.format(epoch, train_loss[-1], val_loss_list[-1]))
                else:
                    print('epoch:{:<2d} | loss:{:<6.7f}'.format(epoch, train_loss[-1]))
        if type(val_loader) is type(None):
            #torch.save(model, os.path.join(out_dir,'checkpoint.pt'))
            pass
        else:
            model = torch.load(os.path.join(out_dir,best_val_loss['name']))
        plt.plot(loss_list, label='train loss')
        plt.plot(val_loss_list, label='val loss')
        plt.legend()
        plt.savefig(f'{out_dir}/history_loss.png', dpi=500)
        plt.close()
    #         if (len(loss_list) > 1) and (abs(loss_list[-2] - loss_list[-1]) < 1e-9):
    #           break
        # mape, rmse = evaluation(y_test=denormalize(Y_val).flatten(), y_predict=denormalize(np.array(y_[-1])))
        # re = relative_error(y_test=Y_val.flatten(), y_predict=y_[-1], threshold=0.7)
        # return model, y_[-1], mape_list, rmse_list, re
        # torch.save(model, os.path.join(out_dir,'lstm.pt'))
        return model, loss_list, val_loss_list
```

```python
def split_dataset(location, input_features, output_features, window_size, batch_size):
  if location == 'v1':
    data = v1_joined
  if location == 'v2':
    data = v2_joined
  if location == 'y6':
    data = y6_joined
  if location == 'y7':
    data = y7_joined

  v1_train = data[:"2022-12-31"].reset_index()
  v1_val = data["2023-01-01":"2023-01-31"].reset_index()
  v1_test = data["2023-02-01":"2023-03-01"].reset_index()

  input_scaler = StandardScaler()
  output_scaler = StandardScaler()
  # input_scaler = MinMaxScaler()
  # output_scaler = MinMaxScaler()

  input_scaler = input_scaler.fit(v1_train[input_features])
  output_scaler = output_scaler.fit(v1_train[output_features])

  v1_train = pd.DataFrame(data=input_scaler.transform(v1_train[input_features]), columns=input_features)
  v1_val = pd.DataFrame(data=input_scaler.transform(v1_val[input_features]), columns=input_features)
  v1_test = pd.DataFrame(data=input_scaler.transform(v1_test[input_features]), columns=input_features)

  x_train, y_train = make_dataset_dl(v1_train, window_size, 48, input_features, output_features)
  x_val, y_val = make_dataset_dl(v1_val, window_size, 48, input_features, output_features)
  x_test, y_test = make_dataset_dl(v1_test, window_size, 48, input_features, output_features)

  train_ds = PowerDataset(x_train, y_train)
  train_loader = DataLoader(train_ds, batch_size=batch_size)

  val_ds = PowerDataset(x_val, y_val)
  val_loader = DataLoader(val_ds, batch_size=batch_size)

  return train_loader, val_loader, (x_test, y_test), input_scaler, output_scaler
```

```python
def lstm_location_train(location, input_features, output_features, window_size, batch_size, device='cpu'):
  t = datetime.datetime.now()

  t = t.strftime("%Y_%m_%d_%H_%M")

  # out_dir = f'/content/{location}_{output_features[0]}_{t}'
  out_dir = f'/content/{location}_{output_features[0]}'

  train_loader, val_loader, (x_test, y_test), input_scaler, output_scaler = split_dataset(location, input_features, output_features, window_size, batch_size)

  model = RNN_Net(len(input_features), 4, 1, 48, task='regression')

  model, loss, val_loss = train(model, train_loader, val_loader, output_scaler, device=device, task='regression', verbose=10, out_dir=out_dir)

  y_pred = model(torch.from_numpy(x_test).to('cpu')).cpu().detach().numpy().reshape(-1,1)
  y_test1 = y_test.copy().reshape(-1,1)
  y_pred = output_scaler.inverse_transform(y_pred).flatten()
  y_test1 = output_scaler.inverse_transform(y_test1).flatten()

  metric = rmr_score(y_test1, y_pred)

  with open(f'{out_dir}/result.txt',"a") as f:
    f.write(f"metric: {metric}")
    f.write('\n')
    f.write(f"input features: {input_features}")
    f.write('\n')
    f.write(f"output features: {output_features}")
    f.write('\n')
    f.write(f"window size: {window_size}")
    f.write('\n')
    f.write(f"batch size: {batch_size}")
    f.write('\n')

  with open(f'{out_dir}/input_scaler.pkl', 'wb') as handle:
    pickle.dump(input_scaler, handle, protocol=pickle.HIGHEST_PROTOCOL)

  with open(f'{out_dir}/output_scaler.pkl', 'wb') as handle:
    pickle.dump(output_scaler, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
class Config:
  def __init__(self, seq_len, pred_len, individual, enc_in):
    self.seq_len = seq_len
    self.pred_len = pred_len
    self.individual = individual
    self.enc_in = enc_in
```

```python
def dlinear_location_train(location, input_features, output_features, window_size, batch_size, device='cpu', config=None):
  """
    output_features: ['power_consumption', 'solar_generation']
  """
  t = datetime.datetime.now()

  t = t.strftime("%Y_%m_%d_%H_%M")

  # out_dir = f'/content/{location}_dlinear_{t}'
  out_dir = f'/content/{location}_dlinear'

  train_loader, val_loader, (x_test, y_test), input_scaler, output_scaler = split_dataset(location, input_features, output_features, window_size, batch_size)

  model = DLinear(config)

  model, loss, val_loss = train(model, train_loader, val_loader, device=device, task='regression', verbose=10, out_dir=out_dir)

  y_pred = model(torch.from_numpy(x_test).to('cpu')).cpu().detach().numpy().reshape(-1,2)
  y_test1 = y_test.copy().reshape(-1,2)
  y_pred = output_scaler.inverse_transform(y_pred)
  y_test1 = output_scaler.inverse_transform(y_test1)

  power_consumption_truth = y_test1[:, 0]
  solar_generation_truth = y_test1[:, 1]

  power_consumption_pred = y_pred[:, 0]
  solar_generation_pred = y_pred[:, 1]

  consumption_metric = rmr_score(power_consumption_truth, power_consumption_pred)
  generation_metric = rmr_score(solar_generation_truth, solar_generation_pred)

  metric = (consumption_metric + generation_metric) / 2

  with open(f'{out_dir}/result.txt',"a") as f:
    f.write(f"metric: {metric}")
    f.write('\n')
    f.write(f"consumption metric: {consumption_metric}")
    f.write('\n')
    f.write(f"generation metric: {generation_metric}")
    f.write('\n')
    f.write(f"input features: {input_features}")
    f.write('\n')
    f.write(f"output features: {output_features}")
    f.write('\n')
    f.write(f"window size: {window_size}")
    f.write('\n')
    f.write(f"batch size: {batch_size}")
    f.write('\n')

  with open(f'{out_dir}/input_scaler.pkl', 'wb') as handle:
    pickle.dump(input_scaler, handle, protocol=pickle.HIGHEST_PROTOCOL)

  with open(f'{out_dir}/output_scaler.pkl', 'wb') as handle:
    pickle.dump(output_scaler, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
def get_excel_input(input_dir_path, input_dir_ex_path, category, date_str):
    # input
    input_files = list(input_dir_path.glob(f"*{category}*.xlsx"))
    if category == 'battery':
        assert 1 >= len(input_files) >= 0
    else:
        assert 2 >= len(input_files) >= 1
    # extra input
    ex_input_files = list(input_dir_ex_path.glob(f"*{category}*_{date_str}.xlsx"))
    assert 1 >= len(ex_input_files) >= 0
    return input_files, ex_input_files

def get_csv_input(input_dir_path, input_dir_ex_path, category, date_str):
    # input
    input_files = list(input_dir_path.glob(f"*{category}*.csv"))
    assert len(input_files) == 1
    # extra input
    ex_input_files = list(input_dir_ex_path.glob(f"*{category}*_{date_str}.csv"))
    assert 1 >= len(ex_input_files) >= 0
    return input_files, ex_input_files
```

```python
def sarima_predict(df, target_date, location_path):

  """
    df is covered through entire January so I hard-code it in the predict code.
    The .pkl file is loaded for t1 only, to change it for other location, remember to change it to the corresponding .pkl file
  """

  target_date = pd.Timestamp(target_date).tz_localize('japan')

  target_end_date = target_date + 47*datetime.timedelta(minutes=30)

  print(target_date, target_end_date)
  power_demand = pd.DataFrame(df['power_consumption'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'],
                            index= df['power_consumption'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'].index)
  power_generation = pd.DataFrame(df['solar_generation'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'],
                                index = df['solar_generation'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'].index)


  consume_scaler, gen_scaler = MinMaxScaler(), MinMaxScaler()


  power_generation_train = power_generation['solar_generation'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'].to_numpy().reshape(-1,1)

  power_consumption_train = power_demand['power_consumption'].loc['2023-01-01 00:00:00+09:00':'2023-01-31 23:30:00+09:00'].to_numpy().reshape(-1,1)

  consume_scaler.fit(power_consumption_train)
  gen_scaler.fit(power_generation_train)
    # Load the ARIMA model from the .pkl file
  with open(f'./{location_path.name}_power_consumption.pkl', 'rb') as f:
      arima_consumption_model = pickle.load(f)

  # Make predictions using the loaded model
  predictions_consumption = arima_consumption_model.predict(start=target_date.isoformat(), end=target_end_date.isoformat(), dynamic=False)


  with open(f'./{location_path.name}_power_generation.pkl','rb') as f:
    arima_generation_model = pickle.load(f)

  predictions_generation = arima_generation_model.predict(start=target_date.isoformat(), end=target_end_date.isoformat(), dynamic=False)

  # Predictions for power_generation
  power_generation_result = gen_scaler.inverse_transform(
      predictions_generation.to_numpy().reshape(-1,1)
  )

  # power_consumption after scaling
  power_demand_result = consume_scaler.inverse_transform(
      predictions_consumption.to_numpy().reshape(-1,1)
  )
  return power_demand_result.flatten(), power_generation_result.flatten()
```

```python
def lstm_uni_predict(model, _df, input_features, input_scaler, output_scaler, device):
  df = _df[input_features]
  df = pd.DataFrame(input_scaler.transform(df), columns=df.columns)
  X = torch.from_numpy(df[input_features].values).float().to(device)
  X = torch.unsqueeze(X, 0)
  model.to(device)
  model.eval()
  with torch.no_grad():
    y = model(X)
  y = np.reshape(y.cpu().numpy(), (-1, 1))
  y = output_scaler.inverse_transform(y)
  return y
```

```python
def dlinear_predict(model, _df, input_fetures, input_scaler, output_scaler, device):
  df = _df[input_features]
  df = pd.DataFrame(input_scaler.transform(df), columns=df.columns)
  X = torch.from_numpy(df[input_features].values).float().to(device)
  X = torch.unsqueeze(X, 0)
  model.to(device)
  model.eval()
  with torch.no_grad():
    y = model(X)
  y = np.reshape(y.cpu().numpy(), (-1, 2))
  y = output_scaler.inverse_transform(y)
  power_consumption = y[:, 0]
  solar_generation = y[:, 1]
  return power_consumption, solar_generation
```

```python
def evaluate():
    # evaluate for each location
    metric_generation_list = []
    metric_demand_list = []

    for location_path in tqdm(list(dir_path.glob('*'))):

        input_dir_path = location_path / input_dir_name
        input_dir_ex_path = location_path / input_dir_ex_name
        eval_dir_y_path = location_path / eval_dir_y_name

        current_date = start_date
        while current_date <= end_date:


            date_str = current_date.strftime('%Y-%m-%d')
            print(f"predicting for location: {location_path.name}, date: {date_str}")

            # prepare the input data (EXCEL)
            # solar input
            solar_input_files, solar_ex_input_files = get_excel_input(
                input_dir_path, input_dir_ex_path, 'solar', date_str
            )
            # surplus input
            surplus_input_files, surplus_ex_input_files = get_excel_input(
                input_dir_path, input_dir_ex_path, 'surplus', date_str
            )
            # battery input
            battery_input_files, battery_ex_input_files = get_excel_input(
                input_dir_path, input_dir_ex_path, 'battery', date_str
            )
            # prepare input data (CSV)
            # solar
            solar_csv_input_files, solar_csv_ex_input_files = get_csv_input(
                input_dir_path, input_dir_ex_path, 'solar', date_str
            )
            # cloud
            cloud_csv_input_files, cloud_csv_ex_input_files = get_csv_input(
                input_dir_path, input_dir_ex_path, 'cloud', date_str
            )
            # weather
            weather_csv_input_files, weather_csv_ex_input_files = get_csv_input(
                input_dir_path, input_dir_ex_path, 'weather', date_str
            )

            # your prediction method
            power_generation_pred, power_demand_pred = predict(
                location_path, date_str,
                solar_input_files, solar_ex_input_files, surplus_input_files, surplus_ex_input_files,
                battery_input_files, battery_ex_input_files, solar_csv_input_files, solar_csv_ex_input_files,
                cloud_csv_input_files, cloud_csv_ex_input_files, weather_csv_input_files, weather_csv_ex_input_files
            )

            # get truth data
            power_generation_truth_path = eval_dir_y_path / f"solar_{date_str}.csv"
            power_generation_truth_df = pd.read_csv(power_generation_truth_path, header=None).fillna(0)
            assert not power_generation_truth_df.iloc[:, 0].hasnans
            power_generation_truth = power_generation_truth_df.iloc[:, 0].values.tolist()

            power_demand_truth_path = eval_dir_y_path / f"demand_{date_str}.csv"
            power_demand_truth_df = pd.read_csv(power_demand_truth_path, header=None).fillna(0)
            assert not power_demand_truth_df.iloc[:, 0].hasnans
            power_demand_truth = power_demand_truth_df.iloc[:, 0].values.tolist()

            # get metric for 1 day
            metric_generation = rmr_score(actual=np.array(power_generation_truth), predict=np.array(power_generation_pred))
            metric_demand = rmr_score(actual=np.array(power_demand_truth), predict=np.array(power_demand_pred))

            metric_generation_list.append(metric_generation)
            metric_demand_list.append(metric_demand)

            # increment current_date
            current_date += datetime.timedelta(days=1)

    # get final metric (average for each day)
    power_generation_metric = np.average(np.array(metric_generation_list))
    power_demand_truth_metric = np.average(np.array(metric_demand_list))
    return (power_generation_metric + power_demand_truth_metric) / 2
```
